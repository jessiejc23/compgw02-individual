```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

train = pd.read_csv("train.csv")
```

```python
train.insert(0, 'Imps', 1)
train['OS'],train['browser'] = train['useragent'].str.split('_', 1).str
```

```python
def ad_stats1():
    featurelist=['weekday','hour','region','city','adexchange','OS','browser']
    df = train[['advertiser','payprice','click','weekday','hour','region','city','adexchange','OS','browser']]
    df_ad = df[df['advertiser'].isin([2259])]
   
    for _, f in enumerate(featurelist):
        fig, ax = plt.subplots(figsize=(6,3))
        df2 = df_ad[['payprice','advertiser',f, 'click']]
        df3 = df2.groupby([f, 'advertiser']).sum()
        df3['eCPC'] = df3['payprice']/df3['click']
        df4 = df3.drop(['payprice','click'],axis = 1)
        df4.unstack(level=1).plot.bar(ax=ax)
        plt.show()
```

```python
ad_stats1()
```
