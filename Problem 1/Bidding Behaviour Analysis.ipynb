```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gc

train = pd.read_csv("train.csv")
```

```python
train.insert(0, 'Imps', 1)
train['eCost'] = train['click']*train['payprice']

train['OS'],train['browser'] = train['useragent'].str.split('_', 1).str
```

```python
def ad_stats1():
    
    featurelist=['weekday','hour','region','city','adexchange','OS','browser']
    df = train[['advertiser','payprice','weekday','hour','region','city','adexchange','OS','browser']]
    df_ad = df[df['advertiser'].isin([1458,2259,2261])]
    for _, f in enumerate(featurelist):
        fig, ax = plt.subplots(figsize=(6,3))
        df2 = df_ad[['payprice','advertiser',f]]
        df3 = df2.groupby([f, 'advertiser']).sum()
        df3.unstack(level=1).plot(ax=ax)
        plt.show()
```

```python
ad_stats1()
```
